```python
import pandas as pd
df = pd.read_csv("C:/Users/yoges/Downloads/hr_analytics_1000_data.csv")
print(df)
df.head(5)
```

```python
df.info()
```

```python
df["join_date"] = pd.to_datetime(df["join_date"], errors = "coerce")
df["exit_date"] = pd.to_datetime(df["exit_date"], errors = "coerce")
```

```python
df["last_promotion_date"] = df["last_promotion_date"].fillna(0).astype(int)
df["exit_reason"] = df["exit_reason"].fillna('Active')
```

```python
#Add new column (feature engineering)
# // 365  Days to years convert
```

```python
df["experience_years"] = ((pd.Timestamp("today") - df["join_date"]).dt.days // 365)
```

```python
#Attrition flag for every employees
# 0 -> Active employee
#1  -> Attrited Employee
```

```python
import numpy as np
df["is_attrited"] = np.where(df["exit_date"].isna(), 0, 1)
```

```python
df = df[(df["age"] >= 18) & (df["age"]) <= 65]
df = df[df["salary"] > 0]
```

```python
df.drop_duplicates(subset = "emp_id", inplace = True)
```

```python
df["avg_working_hours"] = df["avg_working_hours"].clip(lower=6, upper = 12)
```

```python
df_clean = df.copy()
df_clean.head()
```

```python
#Overworked But Underload
```

```python
overworked = df_clean[
    (df_clean["avg_working_hours"] > 9.5) &
    (df_clean["salary"] < df_clean["salary"].median())
]
```

```python
#Identified early-career employees at high attrition risks

early_warning = df_clean[
    (df_clean["avg_working_hours"] > 9) &
    (df_clean["performance_rating"] <= 2)
]
```

```python
df_clean.to_csv("hr_analytics_cleaned.csv", index = False)
```
